3. ¿cuál es la probabilidad de obtener un número par y primo en un lanzamiento de dado?

```python
s = {1, 2, 3, 4, 5, 6}
e_1 = {2, 4, 6}
e_2= {2, 3, 5}
e = set.intersection(e_1, e_2) #cuando es "y" es intereseccion y cuando es "o" es union.
c_n1 = len(set.intersection(e, s))
c_d1 = len(s)
Fraction(c_n1, c_d1)
```

```python
def Pr(event:set, space:set) -> Fraction:
    return Fraction(len(set.intersection(event,space)), len(space))
```

4. El método de Montecarlo, nos permite aproximar una probabilidad con base a la repetición continua de un evento. 
¿Cuál es la probabilidad de obtener un número par en un lanzamiento de dados?

```python
events = [2, 4, 6]
space = [1, 2, 3, 4, 5, 6]
frec = []
trials = list(range(1, 1001))
for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(space) in events: 
            total = total + 1
    frec.append(total/trial)
```

```python
plt.plot(trials, frec)
plt.xlabel("Trials")
plt.ylabel("Relative Frecuence")
plt.axhline(y=0.5 , color="tomato")
plt.grid(True)
plt.show()
```
