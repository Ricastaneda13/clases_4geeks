# El modelo Bernoulli se aplica a un random variable que puede asumir sólo los valores. Para este modelo tenemos: 
- $E(X)=p$
- $V(X)=pq$

# Donde p es la probabilidad de éxito  y q es la probabilidad de fracaso. $E(X)$ la esperanza de la variable aleatoria X y $v(X)$ significa la varianza de la variable aleatoria de X

El Modelo binomial puede considerarse como la suma de n variables bernulli independientes una de otra donde se cumpla lo siguiente:

1. Debe haber un número fijo de pruebas repetidas. Estadísticamente independientes
2. Cada prueba debe ser una variable de Bernuoulli. Es decir, puede resultar en un éxito o un fracaso
3. Todas las pruebas deben tener identica probabilidad de éxito

El modelo binomial es discreto y tiene: 
- $E(X)=np$
- $V(X)=npq$

Donde n es el número de ensayos o pruebas, p es la probabilidade de éxito y q la probabilidad de fracaso.


Una moneda es lanzada 10 veces. ¿cuál es la probabilidad de obtener 8 o más caras?

```python
n = 10
p = 0.5
# estamos buscando p > 8
prob = 1 - binom.cdf(6, n, p) 
#cdf = es la función de distribución acumulada que indica P(X =0) + P(X=1) + 
prob
```

Se sabe que el 90% de los trabajadores de una ciudad industrial son miembros de un sindicato. En un estudio de la situación laboral de la ciudad, un economista escoger al azar una muestra de 15 obreros para entrevistas. ¿cuál es la probabilidad de que la muestra contenga 12 o menos miembros de sindicato?

```python
n = 15
p = 0.9

prob =  binom.cdf(12, n, p)
# P(X) <= 12) 
prob
```

Muchos experimentos no ocurren  como un resultado definido de pruebas sino como puntos de tiempol, espacio o volumen al azar. El modelo Poisson es adecuado en estos casos si: 
1. EL numero de ocurrencias del hecho es independiente de una unidad
2. El valor esperado de la variable es proporcional al tamaño de la unidad especificada
3. La probabilidad de más de una ocurrencia del hecho en una unidad especificada muy pequeña es despreciable

Para este módelo tenemos: 
- $E(X)=\lambda$
- $V(X)=\lambda$

Donde $\lambda$ es igual al promedio de ocurrencias por unidad especificada. La distribución Poisson es discreta

En un callcenter ocurren 4 llamadas por minuto ($\lambda$)= 4 ¿Cual es la probabilidad de que no ocurran más de 2 llamadas durante los 2 minutos siguientes?

```python
n = 2
plambda = 8

# P(X<=2) es decir P(X=0)+ P(X)=1 + P(X=2)

prob = poisson.cdf(n, plambda)
prob
```

```python

# la P(X>4) es decir P(X=4) + P(X=5) + P(X=6)
# p(x>=4) = 1 - p(X<=3) = 1 - P(X=0) - P(X=1) - P(X=2) - P(X=3)

n = 3 #numero de exitos
plambda = 8 # ocurrencias en una unidad especificada 4 llamadas por minuto (2 minutos + 4 llamadas) = 8 llamadas
prob = 1 - poisson.cdf(n, plambda)
prob
```

Un fabricante de tejidos de lana afirma que el promedio de defectos en sus productos es 1 por  2 yardas cuadradas. Se extrajo una yarda cuadrada al azar y tenia 3 defectos. ¿cuál es la probabilidad de que obtener 3 o más defectos en cualquier yarda cuadrada si la afirmación del fabricante fuera validad?

```python
plambda = 0.5
n = 3

# P(X>=3) = P(X=3) + P(X=4) + P(X=5)...
# P(X>=3) = 1 - P(X<=2 = 1- p(X=0) - p(X=1) - p(X=2)


prob = 1 - poisson.cdf(n, plambda)
prob
```

El modelo de probabilidad más frecuentemente usado en el análisis económico y comercial es la distribución normal, que puede expresarse en las formas general y estandarizada. Se dice que una variable aleatoria X tiene una distribución normal si es continua, si está en el rango del menos infinito al infinito y si puede expresarse por su media $\mu$ y su desviacion estandar $\sigma$

```python
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x)
plt.plot(x, pdf, c="tomato")
plt.title("PDF Distribución normal")
plt.xlabel("Value")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 1000)
cdf = norm.cdf(x)
plt.plot(x, cdf, c="tomato")
plt.title("cdf distribución normal")
plt.xlabel("Value")
plt.ylabel("probability")
plt.grid(True)
plt.show()
```
